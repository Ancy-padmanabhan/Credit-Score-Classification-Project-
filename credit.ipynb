```python
# Map the Credit_Mix to a new label or numeric values
credit_map={"Good":2,"Standard":1,"Bad":0}
df['Credit_Mix'] = df['Credit_Mix'].replace(credit_map)
```

```python
# Map the Credit_Score to a new label or numeric values
credit_score_mapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
df['Credit_Score'] = df['Credit_Score'].replace(credit_score_mapping)
```

```python
#Map the occupation data to numeric type
occupation_mapping = {
    'Lawyer': 1,
    'Architect': 2,
    'Engineer': 3,
    'Scientist': 4,
    'Mechanic': 5,
    'Accountant': 6,
    'Developer': 7,
    'Media_Manager': 8,
    'Teacher': 9,
    'Entrepreneur': 10,
    'Doctor': 11,
    'Journalist': 12,
    'Manager': 13,
    'Musician': 14,
    'Writer': 15
}

df['Occupation'] = df['Occupation'].replace(occupation_mapping)
```

```python
#Mapping Payment behaviour to numeric data
category_mapping = {
    'Low_spent_Small_value_payments':1,
    'High_spent_Medium_value_payments':2,
    'Low_spent_Medium_value_payments': 3,
    'High_spent_Large_value_payments': 4,
    'High_spent_Small_value_payments': 5,
    'Low_spent_Large_value_payments': 6
}
df['Payment_Behaviour'] = df['Payment_Behaviour'].replace(category_mapping)
```

```python
#Mapping Payment_of_Min_Amount to numeric value
Payment_of_Min_Amount_map={"NM":2,"Yes":1,"No":0}
df['Payment_of_Min_Amount'] = df['Credit_Mix'].replace(Payment_of_Min_Amount_map)
```

```python
# Exclude "No Data" and "Not Specified" from consideration
filtered_df = df[~df['Type_of_Loan'].isin(["No Data", "Not Specified"])]

# Get the top 50 most frequent loan types
top_50_loan_types = filtered_df['Type_of_Loan'].value_counts().head(50).index

# Replace non-top 50 loan types with "Other"
df['Type_of_Loan'] = df['Type_of_Loan'].apply(lambda x: x if x in top_50_loan_types else 'Other')

# Print the updated 'Type_of_Loan' column value counts to verify
updated_loan_type_counts = df['Type_of_Loan'].value_counts()

```

```python
#Print Count of top 50 type of loan data
updated_loan_type_counts
```

```python
# Calculate the mean Credit_Score for each Type_of_Loan
loan_type_mean_credit_score = df.groupby('Type_of_Loan')['Credit_Score'].mean()

# Map the mean Credit_Score back to the original DataFrame
df['Type_of_Loan'] = df['Type_of_Loan'].map(loan_type_mean_credit_score)
df.head()
```

```python
# Calculate the mean Age for each Occupation
mean_age_by_occupation = df.groupby('Occupation')['Age'].mean()

# Print the mean Age for each Occupation
print("Mean Age by Occupation:")
print(mean_age_by_occupation)

```

```python
# Update Age values less than 18 with the mean age for their respective Occupation as the legal age to apply for the credit card or loan is 18.
df.loc[df['Age'] < 18, 'Age'] = df.loc[df['Age'] < 18, 'Occupation'].map(mean_age_by_occupation)
df['Age'].value_counts()
```

```python
# List of relevant columns for the boxplot
relevant_columns = [
    'Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
    'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
    'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries',
    'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_EMI_per_month',
    'Amount_invested_monthly', 'Monthly_Balance'
]

# Drop any non-numeric columns (if any)
numeric_df = df[relevant_columns].select_dtypes(include='number')

# Number of columns to plot
num_plots = len(numeric_df.columns)
cols = 3  # Number of columns for the subplot grid
rows = (num_plots + cols - 1) // cols  # Calculate the number of rows needed

# Create a larger figure for subplots
plt.figure(figsize=(cols * 6, rows * 4))

# Loop through each relevant column and plot a boxplot in subplots
for i, column in enumerate(numeric_df.columns, 1):
    plt.subplot(rows, cols, i)
    plt.boxplot(numeric_df[column].dropna(), vert=True,)
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Value Range')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
